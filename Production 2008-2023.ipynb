```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
file_path = pd.read_csv("DAUPSA.csv")
new_df = pd.DataFrame ({
    "Country": ["USA"] * len(file_path),
    "Date": file_path["DATE"],
    "Production (Thousands)": file_path["Domestic Production"]
})

new_df
```

```python
plt.figure(figsize=(25, 6))
plt.plot(new_df['Date'], new_df["Production (Thousands)"], marker='o', linestyle='-', color='blue')
plt.title('Domestic Production Over Time')
plt.xlabel('Dates')
plt.ylabel('Domestic Production (Thousands)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join("Data Output", 'domestic_production_plot.png'))
plt.show()
```

```python
plt.figure(figsize=(25, 6))
plt.bar(new_df['Date'], new_df["Production (Thousands)"], color='blue')
plt.title('Domestic Production Over Time (Bar Plot)')
plt.xlabel('Dates')
plt.ylabel('Domestic Production (Thousands)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(os.path.join("Data Output", 'domestic_production_bar_plot.png'))
plt.show()
```
